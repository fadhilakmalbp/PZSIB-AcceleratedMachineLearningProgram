```python
import numpy as np
import pandas as pd
```

### 1. Create the following Pandas DataFrame

```python
df = pd.DataFrame({
    "name": ["Anna","Dane","David","Kevin","Joe","Rose"],
    "weight": [45.2,50.5,62.1,64,59,np.nan],
    "heigth": [155,160,162,170,167,159],
    "birth_year": [2000,1997,1996,1987,1988,2003],
    "gender": ["female","male","male","male","male","female"],
    "city": ["New York","Chicago","Austin","New York",np.nan,"New York"]
})
df
```

### 2. Create another Pandas DataFrame as the following and merge it with the previous DataFrame (key=name and nick name). Make sure that no rows are missing from the original DataFrame

```python
df1 = pd.DataFrame({
    "nick_name": ["Anna", "David", "Dane", "Rose"],
    "full_name": ["Anna Levis", "David John", "Dane Sebastian", "Rose Kimberly"]
    })
df1
```

```python
df.merge(df1, how = "outer", left_on = "name", right_on = "nick_name")
```

### 3. Drop nick_name column and fill null values in the full_name from the merged DataFrame with their original name column.

```python
df3 = df.merge(df1, how = "outer", left_on = "name", right_on = "nick_name").drop(columns = "nick_name")
df3.iat[3,6] = "Kevin"
df3.iat[4,6] = "Joe"
df3
```

### 4. Create a new ‘bmi’ column calculated based on the following formula --> *BMI (kg/m^2) = weight / (height/100)^2*

```python
df3["bmi"] = df3["weight"] / (df3["heigth"] / 100)**2
df3
```

### 5. Create a new ‘bmi_class’ column following the definition in the table below. 

```python
df3["bmi_class"] = pd.cut(df3["bmi"],
                         bins = [0, 18.5, 25, 30, 35, 40, 100],
                         labels = ["Underweight","Normal weight","Overweight","Obesity class 1","Obesity class 2","Obesity class 3"])
df3
```
